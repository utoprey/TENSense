```python
analysis_direction = os.getcwd()
data_dir = 'PerceptMapper/data/'
os.chdir('../' + data_dir)
data_dir = os.getcwd()

files_not_logs = []
for files in os.walk(data_dir):
    flag = True

    for file in files[2]:
        if '.csv' in file and flag:
            files_not_logs.append(files[0])
            flag = False
files_log = []
for files in os.walk(data_dir):
    if files[0] not in files_not_logs:
        if files[0].split('\\')[-1] != 'data':
            files_log.append(files[0])

os.chdir(analysis_direction)
print(files_not_logs)
print(files_log)

# os.chdir('../' + savedir)
# savedir = os.getcwd()
# os.chdir(analysis_direction)
```

```python
# Vyacheslav Kudryavsky_R000_RadioCheckSlider

csvs_keys = ['Picture', 'Time',
             'naturalSlider', 'painSlider', 'phantomSlider', 'motorSlider', 'tempSlider', 'maintactileSlider']
additional = ['Vibration', 'Twitching', 'Intention to Make Movement', 'Touch', 'Squeezing', 'Puncture', 'Prickling',
              'Electric Current', 'Hit', 'Pulsation', 'Tickling', 'Itching', 'Tingling', 'Numbness', 'Muscle contraction', 'Was Felt']
for file_dir in files_log:
    print(file_dir)
    mapper = {}
    for name in csvs_keys:
        mapper[name] = []
    for name in additional:
        mapper[name] = []

    pic_n = 0
    for files in os.walk(file_dir):
        #print(files[2])
        for file in files[2]:
            if 'RadioCheckSlider' in file.split('_')[-1]:
                with open(file_dir + '/' + file, "r") as stream:
                    pic_n += 1
                    for line in stream:
                        line = line.split('  ')[-1].split('\n')[0].split(':')
                        if 'Sensation0' not in line:
                            line[-1] = float(line[-1][1:])
                        for name in csvs_keys:
                            if name == line[0]:
                                mapper[name].append(line[-1])
                    mapper['Picture'].append(pic_n)
                    mapper['Time'].append('04:00:00')

                    for name in additional:
                        mapper[name].append(None)
                    stream.close()
    print(file_dir + '/' + file.split('_')[0] + '.csv')
    pd.DataFrame(mapper).to_csv(file_dir + '/' + file.split('_')[0] + '.csv', index=False)
```

```python
for name in csvs_keys:
    print(f'{name} : {len(mapper[name])}')
for name in additional:
    print(f'{name} : {len(mapper[name])}')
```
